# **Gráficos**

### **Análise por % alunos Aprovados e Reprovados**

```python
dados.head(20)
```

```python
dados = dados.sort_values('Notas', ascending=False)
dados = dados.reset_index(drop=True)
dados.head(20)
```

```python
dados.head(20)
print(dados.columns)
```

```python
total_aprovado_reprovado = dados['Situação'].value_counts()
total_aprovado_reprovado
```

```python
aprovados = dados[dados['Situação'] == 'Aprovado'].shape[0]
reprovados = dados[dados['Situação'] == 'Reprovado'].shape[0]
```

```python
total_alunos = aprovados + reprovados
porcentagem_aprovados = (aprovados / total_alunos) * 100
porcentagem_reprovados = (reprovados / total_alunos) * 100
```

```python
labels = ['Aprovados', 'Reprovados']
sizes = [porcentagem_aprovados, porcentagem_reprovados]
```

```python
labels = ['Aprovados', 'Reprovados']
sizes = [porcentagem_aprovados, porcentagem_reprovados]
```

```python
colors = ['Purple', '#C8A2C8']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.axis('equal')
plt.title('Porcentagem de Alunos Aprovados e Reprovados')

plt.show()
```

# **Criando Colunas**

### **Atividade Extracurricular: 40% da nota atual**

```python
dados.head(20)
```

```python
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x*0.4)
dados
```

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados
```

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x: True if x >= 7 else False)
dados
```

```python
dados['Aprovado_final'] = dados['Aprovado_final'].replace({True: 'Aprovado', False: 'Reprovado'})
print(dados.to_string(index=False))
```

```python
dados.query('Aprovado_final == "Reprovado" | Aprovado_final == "Aprovado"')
```

```python
selecao = (dados['Situação'] == 'Reprovado') & (dados['Aprovado_final'] == 'Aprovado')
resultado = dados.loc[selecao]
resultado.head()
```
